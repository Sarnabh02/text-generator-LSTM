```python
# Defining inputs and Outputs
seq_length = 10
input_sequences = []
target_chars = []
print(len(data_indices))
print(seq_length)
print(len(data_indices) - seq_length)
diff = len(data_indices) - seq_length
for i in range(diff):
    input_sequences.append(data_indices[i:i + seq_length])
    print(data_indices[i:i + seq_length])
    target_chars.append(data_indices[i + seq_length])
    print(data_indices[i + seq_length])


    

# Convert input & outputs to tensors, so the model can interpret them
input_sequences = torch.tensor(input_sequences, dtype=torch.long)
target_chars = torch.tensor(target_chars, dtype=torch.long)

print(input_sequences)

print(target_chars)
```

```python
class LSTMLanguageModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LSTMLanguageModel, self).__init__()
        self.hidden_size = hidden_size
        self.embedding = nn.Embedding(input_size, hidden_size)
        self.lstm = nn.LSTM(hidden_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, input_seq):
        embedded = self.embedding(input_seq)
        output, _ = self.lstm(embedded)
        prediction = self.fc(output[:, -1, :])
        return prediction
```

```python

input_size = len(chars)
hidden_size = 384 
output_size = len(chars)
learning_rate = 0.1
num_epochs = 1000

# Initialization
model = LSTMLanguageModel(input_size, hidden_size, output_size)
loss_f = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

# Training 
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(input_sequences)
    loss = loss_f(outputs, target_chars)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}")
```
